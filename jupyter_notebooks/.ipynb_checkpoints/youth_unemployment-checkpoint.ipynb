1. Download data
2. Explore data
3. Extract/analyse data
4. Visualize data

## Step 1 : Download data

First, we download a copy of "*API_SL.UEM.1524.NE.ZS_DS2_en_excel_v2_1886728.xls*" from Github (https://github.com/soda-lab/data-registry/) with python code. The data initially published from the Unemployment, youth total report by International Labour Organization (ILO) (https://ilo.org/). 

**The International Labour Organization** : ILO is a United Nations agency whose mandate is to advance social and economic justice through setting international labour standards.


### Download data with python

```python
import pandas as pd

repository = 'https://github.com/soda-lab/data-registry/'
path = 'blob/main/original_data/'
dataset = 'API_SL.UEM.1524.NE.ZS_DS2_en_excel_v2_1886728.xls'
raw = '?raw=ture'

url = repository + path + dataset + raw
xls = pd.ExcelFile(url)
```

```python
print(xls.sheet_names)
```

Data successfully loaded to the Jupyter notebook. The data includes 3 tables (see above list of tables).

## Step 2: Explore data

Before we start to extract insights from the data, we need to explore the data to identify potential insights. We write python code to print the three data sheets.

```python
df1 = pd.read_excel(xls, sheet_name = 'Data', skiprows=3, header=0)
df1
```

```python
df2 = pd.read_excel(xls, sheet_name = 'Metadata - Countries', skiprows=0, header=0)
df2
```

```python
df3 = pd.read_excel(xls, sheet_name = 'Metadata - Indicators', skiprows=0, header=0)
df3
```

As you can see in the above tables, "Data" table contains "Country Name", "Country Code", "Indicator Name", "Indicator Code", and year from 1960 to 2020. Secondly, "Metadata - Countries" table contains "Country Code", "Region", "IncomeGroup", "SpecialNotes", and "TableName". Lastly "Metadata - Indicators" table contains "INDICATOR_CODE", "INDICATOR_NAME", "SOURCE_NOTE", and "SOURCE_ORGANIZATION".

## Step3: Extract/analyse data

As we interest in comparing the unemployment rate across OECD countries, we'd like to extract "CountryName" column as well as year columns where the column values are not missing. Below python code demonstrates how to extract columns and rows from the data table.

```python
extract_columns = ['Country Name', '1991', '1992', '1993','1994','1995','1996',
                   '1997','1998','1999','2000','2001','2002','2003','2004','2005',
                   '2006','2007','2008','2009','2010','2011','2012','2013','2014',
                   '2015','2016','2017','2018','2019'] 
extract_rows = ['Australia', 'Austria', 'Belgium', 'Canada', 'Switzerland', 'Chile',
                'Colombia', 'Czech Republic', 'Germany', 'Denmark', 'Spain', 'Estonia',
                'Finland', 'France', 'United Kingdom', 'Greece', 
                'Hungary', 'Ireland', 'Iceland', 'Israel', 'Italy','Japan', 
                'Korea, Rep.', 'Lithuania', 'Luxembourg', 'Latvia', 'Mexico',   
                'Netherlands', 'Norway', 'New Zealand', 'Poland', 'Portugal', 'Slovak Republic',
                'Slovenia', 'Sweden', 'Turkey', 'United States']
```

```python
extracted_df = pd.DataFrame(columns=['Country Name'])

rows_no = 0
for index, row in df1.iterrows():
    if row['Country Name'] in extract_rows:
        extracted_df = extracted_df.append(row)
            
extracted_df = extracted_df.dropna(subset=['Country Name'], how='all') # drop rows where it contains meaningless values
extracted_df = extracted_df[extract_columns]
extracted_df
```

As you can see in the above table, we extract youth unemployment rate data for 36 OECD countries between 1991 to 2019. 

## Step4: Visualize data

Here, we'd like to visualize the extracted tables. Below cells show how to visualize the extracted data by using **Matplot** library.

- **Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python

```python
import matplotlib.pyplot as plt

years = ['1991', '1992', '1993','1994','1995','1996',
           '1997','1998','1999','2000','2001','2002','2003','2004','2005',
           '2006','2007','2008','2009','2010','2011','2012','2013','2014',
           '2015','2016','2017','2018','2019']
plt.figure(figsize=(12,7))
x = list(range(1991,2020))
y = list(extracted_df[years].iloc[0])
plt.plot(x, y)

plt.xlabel('Year', fontsize=12)
# Set the y axis label of the current axis.
plt.ylabel('Unemployment rate (%)', fontsize=12)
# Set a title of the current axes.
plt.title('Youth unemployment rate in Australia in 2019', fontsize=17)
plt.ylim((0, 25))   # set the ylim to bottom, top
# Display a figure.
plt.show()
```

```python
import numpy as np

plt.rcdefaults()

fig, ax = plt.subplots(figsize=(9, 16))

# Example data
countries = extract_rows
y_pos = np.arange(len(countries))
unemployment_rate = list(extracted_df['2019'])

ax.barh(y_pos, unemployment_rate, align='center', color='#DA8FCB')
ax.set_yticks(y_pos)
ax.set_yticklabels(countries)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Youth unemployment rate in 2019 (% of total labour force aged 15-24)')
ax.set_title('Youth unemployment rate across 37 OECD countries in 2019')
plt.xlim((0, 40))   # set the ylim to bottom, top
plt.show()
```


As the first figure shows, the percentage of unemployed young people (aged 15-24)  in the labour force decreased from a high of 17.5 percent in 1991 to 8.8 percent in 2008. Unemployment increased to 13.3 percent in 2014 but decreased slightly up until the beginning of 2020 (12 percent). 

Demonstrating the significant variations between countries, second figure describes the youth unemployment rate across 37 OECD countries in 2019. Greece and Spain have the highest unemployment rates in 2019, both in excess of 30 percent while Italy and Turkey are the second highest group with youth unemployment rates of over 25 percent. By contrast, Japan reports a youth unemployment rate of below 10 percent. The average youth unemployment rate across the 37 OECD countries is 13.7percent.
