## Step 1 : Download data

First, we download copy of "*2071.0.55.001 - Census of Population and Housing: Commuting to Work - More Stories from the Census, Commuting Distance by Personal Characteristics*" from Github (https://github.com/soda-lab/data-registry/) with python. The data initially published from the Australian Bureau of Statistics (ABS)

- **Australian Beureau of Statistics (ABS)**: ABS is a national statistical agency in Australia. It provides a wide range of statistical data on economic, population, environmental, and social issues. 

- **2071.0.55.001 - Census of Population and Housing: Commuting to Work - More Stories from the Census, Commuting Distance by Personal Characteristics**: Commuting distance data provides a measurement of the distance travelled between a person's Place of Usual Residence and Place of Work. There are many socio-economic factors that may influence the distance a person commutes to work. This article explores commuting distance data in association with some personal characteristics collected within the 2016 Census of Population and Housing. This has included analysis of commuting distance by age, sex, occupation, industry, income, education and mode of transport. The data presented in this article explores the distance people travel to work from their Place of Usual Residence. [1]

### Download data with python

```python
import pandas as pd

repository = 'https://github.com/soda-lab/data-registry/'
path = 'blob/main/original_data/'
dataset = '2071055001%20-%20commuting%20distance%20by%20personal%20characteristics.xls'
raw = '?raw=ture'

url = repository + path + dataset + raw
xls = pd.ExcelFile(url)
```

```python
# print sheet name
print(xls.sheet_names)
```

Data successfully loaded to the Jupyter notebook. The data includes "Contents", "Table 1 to 8", and "Explanatory Notes".

## Step3: Extract/analyse data

Since every table has a different column to measure commuting distances, we consider extracting and analyze every table in the dataset. Below cells show how to extract data with python code and the results of data extraction.

### Table 1 Commuting distance by Age

```python
extracted_df1 = pd.DataFrame(columns=['Australian Statistical Geography Standard (ASGS) Description'])
defined_columns = ['Australian Statistical Geography Standard (ASGS) Description', 
                   'Age group (AGEP)',
                   'Average commuting distance \n(kilometres)']
defined_geography = ['Australia']
df1 = df1[defined_columns] # extract data by columns

for index, row in df1.iterrows():
    if row['Australian Statistical Geography Standard (ASGS) Description'] in defined_geography:
        extracted_df1 = extracted_df1.append(row)
        
extracted_df1
```

As you can see from the above table, the average commuting distance by age group, age group (30-39 years) has the greatest average commuting distance (16.59km) following by the age group (50-59 years) (16.42km). The average commuting distance of total population in Australia is 15.98km in 2016.

### Table 1 Commuting distance by State and territory

In table 1, we can also extract data by state and territory

```python
extracted_df1_2 = pd.DataFrame(columns=['Australian Statistical Geography Standard (ASGS) Description'])
defined_columns = ['Australian Statistical Geography Standard (ASGS) Description', 
                   'Age group (AGEP)',
                   'Average commuting distance \n(kilometres)']
geography_groups = ['New South Wales',
                   'Victoria',
                   'Queensland',
                   'South Australia',
                   'Western Australia',
                   'Tasmania',
                   'Northern Territory',
                   'Australian Capital Territory']
df1 = df1[defined_columns] # extract data by columns
for index, row in df1.iterrows():
    if row['Australian Statistical Geography Standard (ASGS) Description'] in geography_groups:
        if row['Age group (AGEP)'] == 'Total':
            extracted_df1_2 = extracted_df1_2.append(row) 
extracted_df1_2 = extracted_df1_2.drop_duplicates(subset=['Australian Statistical Geography Standard (ASGS) Description', 'Age group (AGEP)'])
extracted_df1_2
```

As you can see from the above table, the average commuting distance by state and territory. People living in QLD have the greatest average commuting distance (16.6km) while people living in the ACT have the lowest average commuting distance (11.75km) in 2016.

## Step4: Visualize data

Here, we'd like to visualize the extracted tables. Below cells show how to visualize the extracted data by using **Matplot** library.
- **Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

### AVG commuting distance by AGE

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(extracted_df1['Age group (AGEP)'])[:-1]
AVG_distances = list(extracted_df1['Average commuting distance \n(kilometres)'])[:-1]

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#F86051')
plt.xlabel("Age groups", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by age", fontsize=28)
plt.ylim((0, 25))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index - 0.22, value - 0.8, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=18, rotation = 90)
plt.show()
```
