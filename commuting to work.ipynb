## Step 1 : Download data

- All copy of original data files are uploaded in git repository (https://github.com/soda-lab/data-registry/original_data)

```python
import pandas as pd

repository = 'https://github.com/soda-lab/data-registry/'
path = 'blob/main/original_data/'
dataset = '2071055001%20-%20commuting%20distance%20by%20personal%20characteristics.xls'
raw = '?raw=ture'

url = repository + path + dataset + raw
xls = pd.ExcelFile(url)
```

## Step 2: Explore data

```python
# read Table 1, 4 and 5 in the dataset - (Commuting distance by Age, Industry, and Income)

df1 = pd.read_excel(xls, sheet_name = 'Table 1', skiprows=5, header=0)
df2 = pd.read_excel(xls, sheet_name = 'Table 4', skiprows=5, header=0)
df3 = pd.read_excel(xls, sheet_name = 'Table 5', skiprows=5, header=0)

df1 = df1.dropna(subset=['Australian Statistical Geography Standard (ASGS) Code'], how='all') # drop rows where it contains meaningless values
df2 = df2.dropna(subset=['Australian Statistical Geography Standard (ASGS) Code'], how='all') # drop rows where it contains meaningless values
df3 = df3.dropna(subset=['Australian Statistical Geography Standard (ASGS) Code'], how='all') # drop rows where it contains meaningless values
```

## Step3: Extract/analyse data

### Table 1 (Commuting distance by Age)
- Extract data 
    - column : Australian Statistical Geography Standard (ASGS) Description, Age group (AGEP), Average commuting distance (kilometres)
    - rows :  15-19 years, 20-29 years, 35-39 years, 45-49 years, 55-59 years, 65-69 years, 75-79 years, 85-89 years, 85 years and over 


```python
# extract data from table 1

extracted_df1 = pd.DataFrame(columns=['Australian Statistical Geography Standard (ASGS) Description'])
defined_columns = ['Australian Statistical Geography Standard (ASGS) Description', 
                   'Age group (AGEP)',
                   'Average commuting distance \n(kilometres)']
age_groups = ['15-19 years', 
              '20-29 years', 
              '30-39 years', 
              '40-49 years', 
              '50-59 years', 
              '60-69 years', 
              '70-79 years', 
              '80-89 years', 
              '85 years and over']
df1 = df1[defined_columns] # extract data by columns

for index, row in df1.iterrows():
    if row['Australian Statistical Geography Standard (ASGS) Description'] == 'Australia':
        if row['Age group (AGEP)'] in age_groups:
            extracted_df1 = extracted_df1.append(row) 
```

```python
extracted_df1
```

- The above table shows the average commuting distance(km) by Age groups

### Table 1 (Commuting distance by State and territory)
- Extract data 
    - column : Australian Statistical Geography Standard (ASGS) Description, Age group (AGEP), Average commuting distance (kilometres)
    - rows : 'New South Wales','Victoria','Queensland','South Australia','Western Australia','Tasmania','Northern Territory','Australian Capital Territory'

```python
# extract data from table 1

extracted_df2 = pd.DataFrame(columns=['Australian Statistical Geography Standard (ASGS) Description'])
defined_columns = ['Australian Statistical Geography Standard (ASGS) Description', 
                   'Age group (AGEP)',
                   'Average commuting distance \n(kilometres)']
geography_groups = ['New South Wales',
                   'Victoria',
                   'Queensland',
                   'South Australia',
                   'Western Australia',
                   'Tasmania',
                   'Northern Territory',
                   'Australian Capital Territory']
df1 = df1[defined_columns] # extract data by columns

for index, row in df1.iterrows():
    if row['Australian Statistical Geography Standard (ASGS) Description'] in geography_groups:
        if row['Age group (AGEP)'] == 'Total':
            extracted_df2 = extracted_df2.append(row) 
extracted_df2 = extracted_df2.drop_duplicates(subset=['Australian Statistical Geography Standard (ASGS) Description', 'Age group (AGEP)'])
```

```python
extracted_df2
```

- The above table shows the average commuting distance(km) by state and terrotory

### Table 5 (Commuting distance by Income)

- Extract data 
    - column : Australian Statistical Geography Standard (ASGS) Description, Average commuting distance (kilometres), Weekly Income (INCP)
    - rows : (Australian Statistical Geography Standard (ASGS) Description: 'Australian')

```python
# extract data from table 5 (commuting distance by income)

extracted_df4 = pd.DataFrame(columns=['Australian Statistical Geography Standard (ASGS) Description'])
defined_columns = ['Australian Statistical Geography Standard (ASGS) Description', 
                   'Weekly Income (INCP)',
                   'Average commuting distance \n(kilometres)']
geography_group = ['Australia']
df3 = df3[defined_columns] # extract data by columns

for index, row in df3.iterrows():
    if row['Australian Statistical Geography Standard (ASGS) Description'] == 'Australia':
        extracted_df4 = extracted_df4.append(row) 
extracted_df4
```

## Step4: Visualize data

### AVG commuting distance by AGE

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(extracted_df1['Age group (AGEP)'])
AVG_distances = list(extracted_df1['Average commuting distance \n(kilometres)'])

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#F86051')
plt.xlabel("Age groups", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by age", fontsize=28)
plt.ylim((0, 25))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index - 0.22, value - 0.8, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=16)
plt.show()
```
